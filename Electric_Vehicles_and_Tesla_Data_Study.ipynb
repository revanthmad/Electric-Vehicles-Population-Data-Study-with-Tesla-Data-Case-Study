# **Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# @title Electric Vehicle Type

plt.figure(figsize=(10, 6))
df.groupby('Electric Vehicle Type').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2', n_colors=len(df['Electric Vehicle Type'].unique())))
plt.xlabel('Count')
plt.ylabel('Electric Vehicle Type')
plt.title('Electric Vehicle Type Distribution')
plt.show()
```

```python
# @title Trend Plot of Electric Range Improvement Over the Years

df.groupby('Model Year')['Electric Range'].mean().plot(kind = 'line', figsize = (16, 9))

plt.xlabel('Model Year')
plt.ylabel('Average Electric Range (miles)')
plt.title('Trend of Electric Range Improvement Over the Years')
plt.show()
```

```python
# @title Average Electric Range of Top 10 Electric Vehicle Makes

plt.figure(figsize=(10, 6))

top_10_makes = df['Make'].value_counts().nlargest(10).index
avg_range_by_make = df[df['Make'].isin(top_10_makes)].groupby('Make')['Electric Range'].mean()

avg_range_by_make.plot(kind='bar')
plt.xlabel('Make')
plt.ylabel('Average Electric Range')
plt.title('Average Electric Range of Top 10 Electric Vehicle Makes')
plt.show()
```

```python
# @title Electric Range vs. Base MSRP by Electric Vehicle Type

plt.figure(figsize=(16, 9))

groups = df.groupby('Electric Vehicle Type')
for name, group in groups:
    plt.scatter(group['Electric Range'], group['Base MSRP'], label=name)

plt.xlabel('Electric Range')
plt.ylabel('Base MSRP')
plt.title('Electric Range vs. Base MSRP by Electric Vehicle Type')
plt.legend()
plt.show()
```

# **Case Study: Tesla Data**

```python
# @title Creating a Tesla dataset
tesla_df = df.loc[(df['Make'] == 'TESLA') & df['Model']].reset_index(drop=True)
tesla_df.head(10)
```

```python
tesla_df["Electric Vehicle Type"].value_counts()
```

```python
# @title Electric Range vs. Base MSRP by Model

plt.figure(figsize=(10, 6))
for model in tesla_df['Model'].unique():
    subset = tesla_df[tesla_df['Model'] == model]
    plt.scatter(subset['Electric Range'], subset['Base MSRP'], label=model)

plt.xlabel('Electric Range')
plt.ylabel('Base MSRP')
plt.title('Electric Range vs. Base MSRP by Model')
plt.legend()
plt.show()
```

```python
# @title Distribution of Tesla Models by County Plot

tesla_df.groupby(['County', 'Model'])['Model'].count().unstack().plot(kind='bar', figsize = (21, 9), stacked=True)
plt.xlabel('County')
plt.ylabel('Number of Tesla Cars')
plt.title('Distribution of Tesla Models by County')
```

```python
# @title Tesla Registrations Over Time

plt.figure(figsize=(21, 9))

tesla_df['Model Year'].value_counts().sort_index().plot(kind='line')

plt.xlabel('Model Year')
plt.ylabel('Number of Registrations')
plt.title('Tesla Registrations Over Time')
plt.legend()
plt.show()
```

# **Data Analysis**


```python
# @title What is the most common 'Make' of electric vehicles in the dataset?

print(df['Make'].value_counts().idxmax())
```

```python
# @title How many unique cities are represented in the dataset?

print(str(df['City'].nunique()) + " Unique Cities")
```

```python
# @title What is the average Electric Range of the vehicles?

print("The average Electric Range: " + f"{df['Electric Range'].mean():.2f} miles")
```

```python
# @title What are the top 5 cities with the most electric vehicles?

print(df['City'].value_counts().head(5))
```

```python
# @title What is the highest Electric Range in the dataset?

print(f"Highest Electric Range: {df['Electric Range'].max()} miles")
```

```python
# @title How many vehicles are classified as Clean Alternative Fuel Vehicle (CAFV)?

df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
```

```python
print(f"There are {df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'Clean Alternative Fuel Vehicle Eligible'].shape[0]} vehicles eligible for CAFV.")
```

```python
# @title How many vehicles are classified as BEVs (Battery Electric Vehicles)?

print(f"There are {df[df['Electric Vehicle Type'] == 'Battery Electric Vehicle (BEV)'].shape[0]} BEVs in the dataset.")
```

```python
# @title What percentage of vehicles are from the each year?

print(df['Model Year'].value_counts(normalize=True) * 100)
plt.plot(df['Model Year'].value_counts(normalize=True))
plt.xlabel('Model Year')
plt.ylabel('Percentage of Vehicles')
plt.title('Percentage of Vehicles by Model Year')
plt.show()
```

```python
# @title Plot the distribution of Electric Vehicle Types (BEV, PHEV)

df['Electric Vehicle Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), title= "BEV vs PHEV", colormap = 'tab20')
plt.show()
```

```python
# @title How does the average "Electric Range" vary by "Make"?

df.groupby('Make')['Electric Range'].mean().plot(kind='bar', color = 'skyblue', figsize=(21, 9))
plt.xlabel('Make')
plt.ylabel('Average Electric Range')
plt.title('Average Electric Range by Make')
plt.show()
```

```python
# @title Which Model Year has the highest average Base MSRP?

print(f"The Model Year with the highest average Base MSRP is {df.groupby('Model Year')['Base MSRP'].mean().idxmax()}.")
```

```python
# @title Check the correlation factor between Electric Range and Base MSRP

print(f"The correlation between Electric Range and Base MSRP is: {df['Electric Range'].corr(df['Base MSRP']):.2f}")
```
